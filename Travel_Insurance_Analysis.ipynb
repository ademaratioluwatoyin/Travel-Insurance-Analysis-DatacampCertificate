### Travel Insurance Case Study

```python
# importing pandas to load dataset
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('/content/travel_insurance.csv')
data['TravelInsurance'] = data['TravelInsurance'].apply(lambda x: 'Not a customer' if x == 0 else 'Is a Customer')
data.head()
```

```python
# setting figure size 
sns.set(rc={'figure.figsize':(8,5)})

g = sns.histplot(data = data, x = 'Age', hue = 'TravelInsurance', bins = 10, element = 'poly')
g.set(xlabel = 'Age',
      ylabel = 'Frequency',
      title = 'Age Distribution of customers and non-customers')
plt.show()
```

```python
customer_age = data[data['TravelInsurance'] == 'Is a Customer']
g = sns.histplot(data = customer_age, x = 'Age', bins = 10, element = 'poly')
g.set(xlabel = 'Age',
      ylabel = 'Frequency',
      title = 'Age Distribution of customers')
plt.show()
```

```python
customer_age = data[data['TravelInsurance'] == 'Is a Customer']
print('Customers\' Ages and their frequency')
print(customer_age['Age'].value_counts())
```

```python
customer_age['Age'].value_counts().to_frame().plot(kind = 'bar')
plt.show()
```

Most customers are around 34 years old

```python
g = sns.countplot(x = 'Employment Type', data = data, hue = 'TravelInsurance')
g.set(xlabel = 'Employment Type',
      ylabel = 'Count',
      title = 'Employment Type by Travel Insurance')

plt.show()
```

more people in the private sector/self employed are customers

```python
average_income = data.groupby('TravelInsurance')[['AnnualIncome']].mean()/1000
print('Average annual income in thousands')
print()
print(average_income)
```

```python
g = sns.barplot(x = average_income.index, y = average_income.AnnualIncome, data = average_income)
g.set(ylabel = 'Average Annual Income in thousands',
      xlabel = 'Travel Insurance Status',
      title = 'Average Annual Income for Non-customers and Customers')
plt.show()
```

```python
g = sns.histplot(data = data, x = 'AnnualIncome', hue = 'TravelInsurance', bins = 10, element = 'poly')
g.set(xlabel = 'Annual Income',
      ylabel = 'Frequency',
      title = 'Annual Income Distribution of customers and non-customers')
plt.show()
```

```python
customer_annual_income = data[data['TravelInsurance'] == 'Is a Customer']
g = sns.histplot(x = 'AnnualIncome', data = customer_annual_income, bins = 10)
g.set(title = 'Annual Income Distribution of Customers')
plt.show()
```

```python
customer_annual_income = data[data['TravelInsurance'] == 'Not a customer']
g = sns.histplot(x = 'AnnualIncome', data = customer_annual_income, bins = 10)
g.set(title = 'Annual Income Distribution of Non-customers')
plt.show()
```

those who are customers earn more 

```python
g = sns.countplot(x = 'GraduateOrNot', data = data, hue = 'TravelInsurance')
g.set(title = 'Graduation Status of Customers and Non-customers',
      xticklabels = ['Graduate', 'Not a Graduate'])

plt.show()
```

People that travel are mostly Graduate with even more of the customers being graduates

```python
g = sns.countplot(x = 'FamilyMembers', data = data, hue = 'TravelInsurance')
g.set(title = 'Number of family members of travellers')
plt.show()
```

```python
g = sns.countplot(x = 'TravelInsurance', data = data, hue = 'ChronicDiseases')
g.set(title = 'Number of Travellers with or without Chronic Diseases')
plt.show()
```

```python
g = sns.countplot(x = 'EverTravelledAbroad', data = data, hue = 'TravelInsurance')
g.set(title = 'Travellers that have travelled abroad or not')
plt.show()
```
